```python
Q1_ins = df['Insulin'].quantile(0.25)
Q3_ins = df['Insulin'].quantile(0.75)
IQR_ins = Q3_ins - Q1_ins
df = df[(df['Insulin'] >= Q1_ins - 1.5*IQR_ins) & 
        (df['Insulin'] <= Q3_ins + 1.5*IQR_ins)]
df.shape
```

```python
Q1_bmi = df['BMI'].quantile(0.25)
Q3_bmi = df['BMI'].quantile(0.75)
IQR_bmi = Q3_bmi - Q1_bmi
df = df[(df['BMI'] >= Q1_bmi - 1.5*IQR_bmi) & 
        (df['BMI'] <= Q3_bmi + 1.5*IQR_bmi)]
df.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_scaled=df.copy()
df_scaled=df.drop(columns=['Outcome'])
df_scaled[df.columns[:-1]]=scaler.fit_transform(df[df.columns[:-1]])
df_scaled.describe()
```

```python
from sklearn.model_selection import train_test_split
X=df_scaled
y=df['Outcome']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(X_train,y_train)
```

```python
# training score
y_train_pred=clf.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
train_accuracy= accuracy_score(y_train,y_train_pred)
print("Training Accuracy: ",train_accuracy)
```

```python
# testing score
y_pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
test_accuracy= accuracy_score(y_test,y_pred)
print("Training Accuracy: ",test_accuracy)
```

```python
confusion_matrix(y_train,y_train_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import precision_score,recall_score,classification_report
precision_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
